In this Assignment, I visualize population density of LA county in the map. Steps are below:

### Step 1. Load and Clean the Population Data

```python
import pandas as pd
import geopandas as gpd
import re

# Load the population data
population_file = "23_LA_Population.csv"  # Make sure the path is correct
population_data = pd.read_csv(population_file)
population_data
```

```python
# Extract Census Tract numbers
# Extract numeric part (e.g., 'Census Tract 1011.10; Los Angeles County; California' -> '1011.10')
population_data["Census_Tract_Name"] = population_data["Census_Tract_Name"].str.extract(r"Census Tract (\d+\.\d+|\d+)")

# Clean the 'Total population' column
# Remove commas and convert to integers
population_data["Total population"] = population_data["Total population"].str.replace(",", "").astype(int)

# Preview the cleaned data
print(population_data.head(10))

# Save the cleaned data to a new CSV file
cleaned_population_file = "Cleaned_Census_Population.csv"
population_data.to_csv(cleaned_population_file, index=False)

print(f"The cleaned data has been saved to {cleaned_population_file}")
```

### Step 2. Load the LA County Geographic Shapefile

```python
import geopandas as gpd
# Shapefile Path
shapefile_path = "2020_Census_Tracts.shp"

# Load the Shapefile
geo_data = gpd.read_file(shapefile_path)

# Look at the first 5 row of the data
print(geo_data.head())

# Display the structure of the data
print(geo_data.info())
```

### Step 3. Merge Population and Geographic Data

```python
import pandas as pd
import geopandas as gpd


# Load the cleaned population data
population_data = pd.read_csv("Cleaned_Census_Population.csv")

# Load the geographic shapefile
shapefile_path = "2020_Census_Tracts.shp"  # Make sure the path is correct
geo_data = gpd.read_file(shapefile_path)

# Ensure column types match for merging
population_data["Census_Tract_Name"] = population_data["Census_Tract_Name"].astype(str)
geo_data["LABEL"] = geo_data["LABEL"].astype(str)

# Merge the data
merged_data = geo_data.merge(population_data, left_on="LABEL", right_on="Census_Tract_Name", how="inner")

# Convert the merged GeoDataFrame to WGS84 CRS (required for Folium)
merged_data = merged_data.to_crs(epsg=4326)
```

### Step 4. Calculate Population Density

```python
merged_data['Population Density'] = merged_data['Total population'] / (merged_data['ShapeSTAre'] / 1e6)  # Area in km²
```

### Step 5. Create a Static Map

```python
import folium
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1, figsize=(12, 10))
merged_data.plot(
    column="Population Density",
    cmap="YlOrRd",  # Maintain the YlOrRd color scheme
    linewidth=0.8,
    ax=ax,
    edgecolor="0.6",  # Slightly darker edge lines for better visibility
    legend=True
)

# Customize the static plot
ax.set_title("Population Density by Census Tract", fontsize=16)
ax.set_axis_off()  # Remove axis for better visualization

# Save the static map as an image
static_map_file = "Static Population Density Map.png"
plt.savefig(static_map_file, dpi=300, bbox_inches="tight")
plt.show()

print(f"The static map with enhanced contrast has been saved to {static_map_file}")
```
