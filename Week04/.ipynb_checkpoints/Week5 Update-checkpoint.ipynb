```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
from shapely.geometry import Point
import numpy as np

# 
census_population_path = 'Cleaned_Census_Population.csv'
bike_crashes_path = '23_Bicycle_Crashes.csv'
shapefile_path = '2020_Census_Tracts.shp'

# 
census_population = pd.read_csv(census_population_path)
bike_crashes = pd.read_csv(bike_crashes_path)
census_tracts = gpd.read_file(shapefile_path)

# 
census_tracts = census_tracts.to_crs("EPSG:4326")
bike_crashes['geometry'] = bike_crashes.apply(lambda row: Point(row['POINT_X'], row['POINT_Y']), axis=1)
bike_crashes_gdf = gpd.GeoDataFrame(bike_crashes, geometry='geometry', crs="EPSG:4326")

# 
bike_crashes_gdf = gpd.sjoin(bike_crashes_gdf, census_tracts, how="inner", predicate="within")
crash_counts = bike_crashes_gdf.groupby('LABEL').size().reset_index(name='crash_count')

# 
census_population['LABEL'] = census_population['Census_Tract_Name'].astype(str)
merged_data = census_tracts.merge(census_population, on='LABEL', how='left')
merged_data = merged_data.merge(crash_counts, on='LABEL', how='left')
merged_data['crash_count'] = merged_data['crash_count'].fillna(0)
merged_data['Total population'] = merged_data['Total population'].fillna(0)
merged_data['ShapeSTAre'] = merged_data['ShapeSTAre'].fillna(0)

# 
merged_data['population_density'] = merged_data['Total population'] / (merged_data['ShapeSTAre'] / 1e6)
merged_data['crash_rate'] = (merged_data['crash_count'] / merged_data['Total population']) * 1000
merged_data['crash_rate'] = merged_data['crash_rate'].fillna(0)

# 
# 
merged_data = merged_data[merged_data['population_density'] > 0]
merged_data['density_group'] = pd.qcut(merged_data['population_density'], q=3, labels=['Low Density', 'Medium Density', 'High Density'])

# 
grouped_data = merged_data.groupby('density_group')['crash_rate'].mean().reset_index()

# 
plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged_data, x='population_density', y='crash_rate', alpha=0.6)
plt.title('Population Density vs Crash Rate', fontsize=16)
plt.xlabel('Population Density (people/km)', fontsize=14)
plt.ylabel('Crash Rate (crashes per 1000 people)', fontsize=14)
plt.grid(True)
plt.show()

# 
plt.figure(figsize=(8, 6))
sns.barplot(data=grouped_data, x='density_group', y='crash_rate', palette='viridis')
plt.title('Average Crash Rate by Population Density Group', fontsize=16)
plt.xlabel('Population Density Group', fontsize=14)
plt.ylabel('Average Crash Rate (crashes per 1000 people)', fontsize=14)
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.colors as mcolors
import folium
from folium.features import GeoJsonTooltip

# 
census_population_path = 'Cleaned_Census_Population.csv'
bike_crashes_path = '23_Bicycle_Crashes.csv'
shapefile_path = '2020_Census_Tracts.shp'

# 
census_population = pd.read_csv(census_population_path)
bike_crashes = pd.read_csv(bike_crashes_path)
census_tracts = gpd.read_file(shapefile_path)

# 
census_tracts = census_tracts.to_crs("EPSG:4326")
bike_crashes['geometry'] = bike_crashes.apply(lambda row: Point(row['POINT_X'], row['POINT_Y']), axis=1)
bike_crashes_gdf = gpd.GeoDataFrame(bike_crashes, geometry='geometry', crs="EPSG:4326")

# 
bike_crashes_gdf = gpd.sjoin(bike_crashes_gdf, census_tracts, how="inner", predicate="within")
crash_counts = bike_crashes_gdf.groupby('LABEL').size().reset_index(name='crash_count')

# 
census_population['LABEL'] = census_population['Census_Tract_Name'].astype(str)
merged_data = census_tracts.merge(census_population, on='LABEL', how='left')
merged_data = merged_data.merge(crash_counts, on='LABEL', how='left')

# 
merged_data['crash_count'] = merged_data['crash_count'].fillna(0)
merged_data['Total population'] = merged_data['Total population'].fillna(0)
merged_data['ShapeSTAre'] = merged_data['ShapeSTAre'].fillna(0)

# 
merged_data['population_density'] = merged_data['Total population'] / (merged_data['ShapeSTAre'] / 1e6)
merged_data['crash_rate'] = (merged_data['crash_count'] / merged_data['Total population']) * 1000
merged_data['crash_rate'] = merged_data['crash_rate'].fillna(0)

#  crash_rate  0  inf 
filtered_data = merged_data[
    (merged_data['crash_rate'] > 0) & 
    (~merged_data['crash_rate'].isin([np.inf, -np.inf]))
]

#  crash_rate  97% 
crash_rate_threshold = filtered_data['crash_rate'].quantile(0.97)
filtered_data = filtered_data[filtered_data['crash_rate'] <= crash_rate_threshold]

# 
output_scatter_data_path = 'scatter_plot_data.csv'
scatter_data = filtered_data[['LABEL', 'population_density', 'crash_rate']]
scatter_data.to_csv(output_scatter_data_path, index=False)

# 
plt.figure(figsize=(12, 10))
ax = plt.gca()

#  crash_rate  0  inf 
map_data = merged_data.copy()
map_data['mapped_crash_rate'] = map_data['crash_rate'].where(
    map_data['crash_rate'] <= crash_rate_threshold, np.nan
)
map_data['mapped_crash_rate'] = map_data['mapped_crash_rate'].replace([0, np.inf, -np.inf], np.nan)

cmap = plt.cm.OrRd
norm = mcolors.Normalize(vmin=map_data['mapped_crash_rate'].min(), vmax=map_data['mapped_crash_rate'].max())

map_data.plot(column='mapped_crash_rate', cmap=cmap, norm=norm, legend=True,
              legend_kwds={'label': "Crash Rate (crashes per 1000 people)", 'orientation': "vertical"},
              edgecolor="black", linewidth=0.5, ax=ax)

plt.title("Crash Rate by Census Tract", fontsize=16)
plt.xlabel("Longitude", fontsize=12)
plt.ylabel("Latitude", fontsize=12)
plt.show()

# 
m = folium.Map(location=[34.05, -118.25], zoom_start=10, tiles="cartodbpositron")

# Folium 
def style_function(feature):
    crash_rate = feature['properties']['mapped_crash_rate']
    if pd.isna(crash_rate):
        return {'fillColor': 'lightgray', 'color': 'black', 'weight': 0.5, 'fillOpacity': 0.6}
    else:
        rgba_color = cmap(norm(crash_rate))
        hex_color = mcolors.to_hex(rgba_color)
        return {'fillColor': hex_color, 'color': 'black', 'weight': 0.5, 'fillOpacity': 0.6}

#  GeoJSON
geojson = folium.GeoJson(
    data=map_data,
    style_function=style_function,
    tooltip=GeoJsonTooltip(
        fields=["LABEL", "population_density", "crash_rate"],
        aliases=["Census Tract:", "Population Density (people/km):", "Crash Rate (crashes/1000 people):"],
        localize=True
    )
)
geojson.add_to(m)

# 
colormap = folium.LinearColormap(
    colors=[mcolors.to_hex(cmap(norm(v))) for v in np.linspace(map_data['mapped_crash_rate'].min(), map_data['mapped_crash_rate'].max(), 256)],
    vmin=map_data['mapped_crash_rate'].min(),
    vmax=map_data['mapped_crash_rate'].max(),
    caption="Crash Rate (crashes per 1000 people)"
)
colormap.add_to(m)

# 
output_map_path = 'interactive_crash_rate_map_with_legend.html'
m.save(output_map_path)

print(f"The interactive map with legend has been saved to {output_map_path}")
```
